```python
import numpy as np
import matplotlib.pyplot as plt

#inputs

L=10 # m
w=50 #N/m

a=2.5
b=5
c=L-(a+b)

R1=(w*b/L)*(c+b/2)
R2=(w*b/L)*(a+b/2)
l=np.linspace(0,L,200)

X=[]
SF=[]
M=[]

for x in l:
    if x<a:
        sf=R1
        m=R1*x
    elif a<x<(a+b):
        sf=R1-w*(x-a)
        m=(R1*x)-w*(x-a)**2/2
    elif x>(a+b):
        sf=-R2
        m=R2*(L-x)
        
    X.append(x)
    SF.append(sf)
    M.append(m)
    
plt.figure(figsize=(7.5,10),dpi=200)
plt.subplot(2,1,1)
plt.plot(X,SF,color="green")
plt.fill_between(X,SF,color="green",hatch="||",alpha=0.5)
plt.title("Shear force diagram")
plt.xlabel("Length of beam")
plt.ylabel("Shear force in N")
plt.xlim(0,L)

plt.tight_layout(pad=4)
plt.subplot(2,1,2)
plt.plot(X,M,color="red")
plt.fill_between(X,M,color="red",hatch="||",alpha=0.5)
plt.title("Momentum")
plt.xlabel("Length of beam")
plt.ylabel("Momentum")
plt.xlim(0,L)

plt.show()
```
