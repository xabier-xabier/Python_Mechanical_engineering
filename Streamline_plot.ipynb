Streamline Plots

```python
import numpy as np
import matplotlib.pyplot as plt

n = 5

x=np.linspace(-n,n,50)
y=np.linspace(-n,n,50)
X,Y = np.meshgrid(x,y)

U = x**2 + Y-1  # U component of fluid velocity
V = X + Y**2 +1 # V component of velocity

plt.figure(figsize=(6,6))
fluid_stream = plt.streamplot(X,Y,U,V,color=np.sqrt(U**2+V**2), density=[1,1],cmap="jet")


plt.colorbar((fluid_stream).lines)
plt.show()
```
