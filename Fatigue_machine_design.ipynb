Fatigue machine design

```python
import numpy as np
import matplotlib.pyplot as plt

#input
Sm=200
Sa=250

Sut=450
Syt=330
Se=250

plt.figure(figsize=(8,8))
def soderberg(syt,se):
    plt.plot([syt,0],[0,se],label="Soderbeg Line",color="blue")
    plt.legend()
    plt.xlim(0,Sut+50)
    plt.ylim(0,Se+50)
    plt.xlabel(r'$\sigma_m$',fontsize=20)
    plt.ylabel(r'$\sigma_a$',fontsize=20)
    return

def goodman(sut,se):
    plt.plot([sut,0],[0,se],label="Goodman Line",color="red")
    plt.legend()
    return

def gerber(sm,se,sut):
    sm=np.linspace(0,sut,100)
    sa=se*(1-((sm/sut)**2))
    plt.plot(sm,sa,color="green",label="Gerber curve")
    plt.legend()
    return

smean=[100,200,178,190]
samp=[150,198,230,50]
plt.scatter(smean,samp)

soderberg(Syt,Se),goodman(Sut,Se),gerber(Sm,Se,Sut)
plt.show()
```
