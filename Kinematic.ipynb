Kinematic of Slider crank mechanism

```python
import numpy as np
import matplotlib.pyplot as plt
from numpy import pi,sin,cos

r=125 #crank radius
c=250 #connecting rad length
rpm= 500
omega=(2*pi*rpm)/60 #rad/sec

def disp(t):
    d1=r*cos(omega*t)
    d2=np.sqrt(c**2-r**2*sin(omega*t)*sin(omega*t))
    return d1+d2

def vel(t):
    return derivative(disp,t)

def acc(t):
    return derivative(vel,t)

time = np.linspace(0,0.5,300)
theta=omega*time*180/pi

x= disp(time)
v=vel(time)
a=acc(time)

plt.figure(figsize=(7.5,15))
plt.subplot(3,1,1)
plt.plot(theta,x,color="black")
plt.ylabel("Displacement in mm")
plt.xlim(0,360*3)
plt.xlabel("Angle of cranck in degree")
plt.tight_layout(pad=3.0)

plt.subplot(3,1,2)
plt.plot(theta,v,color="green")
plt.ylabel("Velocity m/s")
plt.xlim(0,360*3)
plt.xlabel("Angle of cranck in degree")
plt.tight_layout(pad=3.0)

plt.subplot(3,1,3)
plt.plot(theta,a,color="blue")
plt.ylabel("Acceleration m**2/s")
plt.xlabel("Angle of cranck in degree")
plt.xlim(0,360*3)

plt.show()

```
