```python
import numpy as np
import matplotlib.pyplot as plt
import math

d = np.linspace(1,20,20)
alpha = []
beta = []

for x in d:
    a = math.atan(5/x)
    b = math.atan(7/x)
    alpha.append(a)
    beta.append(b)
    P = 500*np.sin(beta)/np.sin(alpha)
    T = 500*((np.sin(beta)/np.tan(alpha))+np.cos(beta))

plt.plot(d,P,color = 'g',marker = '*',label = 'Force')
plt.plot(d,T,color = 'r',marker = 'o',label = 'Resultant force')
plt.legend()
plt.xlabel("Distance in mm")
plt.ylabel("Force in N")
plt.grid('-')
plt.show()
```
