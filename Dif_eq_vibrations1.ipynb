```python
import numpy as np
import matplotlib.pyplot as plt
from scipy.integrate import odeint

#solution=odeint(funtion,x0,x,arg)

def my_equation(y,x):
    dydx=x**3+(2*x**2)
    
    return dydx

y0=4 #value of y when x=0
x=np.linspace(0,1,20)
y=odeint(my_equation,y0,x,)

plt.plot(x,y)
plt.show()
```
