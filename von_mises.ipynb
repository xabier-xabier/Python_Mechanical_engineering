```python
import numpy as np
import matplotlib.pyplot as plt
from numpy import sin,cos,pi

Syt=200 # Yield strength
S1=150  # Stress condition direction 1
S2=80   # stress condition direction 2

Sv=np.sqrt((S1**2)-(S1*S2)+S2**2)

a=2**0.5*Syt # semi major axis of ellipse
b= (2/3)**0.5*Syt # minor axis

theta=np.linspace(0,360*pi/180,720)

x=a*cos(theta) #original point
y=b*sin(theta)

Sp1=[200,160,80,60]
Sp2=[-80,210,50,40]

rot_ang=pi/4

xd=x*cos(rot_ang)-y*sin(rot_ang)
yd=x*sin(rot_ang)+y*cos(rot_ang)

xp=[Syt,Syt,0,-Syt,-Syt,0,Syt]
yp=[0,Syt,Syt,0,-Syt,-Syt,0]

p1=[-200,200]
p2=[0,200]

px1=[0,0]
px2=[-300,300]

plt.figure(figsize=(7.5,5))
plt.plot(xd,yd,label="Von_Mises_Region")
plt.plot(xp,yp,linestyle="--",label="Max Shear Region")
plt.plot(px1,px2,linestyle="--",color="black")
plt.scatter(p1,p2)
plt.legend()
plt.show()
```
