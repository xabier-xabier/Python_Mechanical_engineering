FFT-Fast Fourier Trasnform

```python
import numpy as np
import matplotlib.pyplot as plt
from scipy import fftpack

x = [1,2,3,4,5]
y = fftpack.fft(x)

z=fftpack.ifft(y)

print(z)
```

```python
import numpy as np
import matplotlib.pyplot as plt
from scipy import fftpack

t =np.linspace(-10,10,200)
dt = t[1]-t[0]

freq = 50
width = 2

g = np.exp(-np.abs(t)/width)*np.sin(2*np.pi*freq*t)

G = fftpack.fft(g)
f =fftpack.fftfreq(g.size,d=dt)

plt.figure(figsize=(15,7.5))
plt.subplot(2,1,1)
plt.plot(t,g)

plt.subplot(2,1,2)
plt.plot(f,abs(G))
plt.show()
```
