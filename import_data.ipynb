```python
import numpy as np
import matplotlib.pyplot as plt
import pandas as pd

eng_data = pd.read_csv("engine_file.csv",nrows=722)

eng_data.head()
```

```python
x1=eng_data["Crank Angle (deg)"]
y1=eng_data["Shear Force Pin_1"]
y2=eng_data["Shear Force Pin_2"]

plt.figure(figsize=(10,7.5))
plt.plot(x1,y1,label="Pin_1 shear force")
plt.plot(x1,y2,label="Pin_2 shear force")
plt.legend(loc="upper right")
plt.xlim(0,720)
plt.show()
```
