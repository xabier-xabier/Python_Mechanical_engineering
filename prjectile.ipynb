Projectile motion

```python
import numpy as np
import matplotlib.pyplot as plt
from numpy import sin,cos,pi

u0=40 # projectile velocity
g=9.81 # gravity
theta=45 #projectile angle
theta2=30

ux=u0*cos(theta*pi/180)
uy=u0*sin(theta*pi/180)

t_total=2*uy/g
t=np.linspace(0,t_total,100)

sx=ux*t
sy=(uy*t)-(0.5*g*t**2)

plt.figure(figsize=(10,7.5))
plt.plot(sx,sy,label="theta=45")
plt.ylim(0,50)
plt.legend()
plt.show()
```
