N-R method

```python
import numpy as np
import matplotlib.pyplot as plt
from scipy.misc import derivative

def f(x):
    return x**3-x-1

def g(x):
    return derivative(f,x,dx=1e-4)


x0=0.5
error=1e-8

while f(x0)>error:
    x1=x0-(f(x0)/g(x0))
    x0=x1
    print(x1)

x=np.linspace(-3,3,30)
plt.plot(x,f(x))
plt.grid(linestyle="--",color="black")
```
