Response of single Degree of freedom

```python
import numpy as np
import matplotlib.pyplot as plt
from numpy import sin,cos,pi
from numpy.linalg import inv

#inputs
M=100
C=20
K=50
w=20
F0=100

end_time=40
time_step=0.01
time=np.arange(0,end_time,time_step)

A=np.array([[M,0],
            [0,1]])

B=np.array([[C,K],[-1,0]])

x=np.array([[3],
            [3]])

F=np.zeros([2,1])

# empty arrays
S=[]
V=[]
T=[]

for t in time:
    F[0]=F0*sin(w*t)
    X = x + (time_step*(inv(A)).dot(F-B.dot(x)))
    x = X
    S.append(X[1])
    V.append(X[0])
    T.append(t)

plt.figure(figsize=(5,2.5),dpi=150)
plt.plot(T,S,label="Displacement",lw=1)
plt.plot(T,V,label="Velocity",lw=1)

plt.legend()
plt.xlim(0,end_time)
plt.show()
    
```

```python
import numpy as np
import matplotlib.pyplot as plt
from numpy import sin,cos,pi
from numpy.linalg import inv

#inputs
M=50
C=10
K=100
w=50
F0=100

end_time=40
time_step=0.01
time=np.arange(0,end_time,time_step)

A=np.array([[M,0],
            [0,1]])

B=np.array([[C,K],[-1,0]])

x=np.array([[3],
            [3]])

F=np.zeros([2,1])

# empty arrays
S=[]
V=[]
T=[]

for t in time:
    F[0]=F0*sin(w*t)
    X = x + (time_step*(inv(A)).dot(F-B.dot(x)))
    x = X
    S.append(X[1])
    V.append(X[0])
    T.append(t)

plt.figure(figsize=(5,2.5),dpi=150)
plt.plot(T,S,label="Displacement",lw=1)
plt.plot(T,V,label="Velocity",lw=1)

plt.yticks(np.arange(-6,8,2))
plt.xticks(np.arange(0,end_time,10))

plt.legend()
plt.xlim(0,end_time)
plt.show()
    
```
