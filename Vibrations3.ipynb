Vibrations of single degree of freedom system

```python
import numpy as np
import matplotlib.pyplot as plt
from scipy.integrate import odeint

def SDOF(X,t):
    m=100
    c=5
    k=50
    F0=100
    w=5
    u=X[0]
    v=X[1]
    
    dudt = v
    dvdt = (F0*np.sin(w*t)/m)-(c*v/m)-(k*u/m)
    
    return dudt, dvdt

time = np.linspace(0,50,1000)
X0= [0,3]

soln = odeint(SDOF,X0,time)

disp =soln[:,0]
velo =soln[:,1]

plt.figure(figsize=(8,5))
plt.plot(time,disp,label="Displacement vs time")
plt.plot(time,velo,label="Velocity vs time")
plt.legend()
plt.xlim(time[0],time[-1])
plt.show()
```
